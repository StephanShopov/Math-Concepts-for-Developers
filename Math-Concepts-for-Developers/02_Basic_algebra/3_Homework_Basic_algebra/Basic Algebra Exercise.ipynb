### Problem 1. Polynomial Interpolation
We know that if we have a set of $n$ data points with coordinates $(x_1; y_1), (x_2; y_2), \dots, (x_n; y_n)$, we can try to figure out what function may have generated these points.

Please note that **our assumptions about the data** will lead us to choosing one function over another. This means that our results are as good as our data and assumptions. Therefore, it's extremely important that we write down our assumptions (which sometimes can be difficult as we sometimes don't realize we're making them). It will be better for our readers if they know what those assumptions and models are.

In this case, we'll state two assumptions:
1. The points in our dataset are generated by a polynomial function
2. The points are very precise, there is absolutely no error in them. This means that the function should pass **through every point**

This method is called *polynomial interpolation* (*"polynomial"* captures assumption 1 and *"interpolation"* captures assumption 2).

It can be proved (look at [Wikipedia](https://en.wikipedia.org/wiki/Polynomial_interpolation) for example) that if we have $n$ data points, there is only one polynomial of degree $n-1$ which passes through them. In "math speak": "the vector spaces of $n$ points and polynomials of degree $n-1$ are isomorphic (there exists a bijection mapping one to the other)".

There are a lot of ways to do interpolation. We can also write the function ourselves if we want but this requires quite a lot more knowledge than we already covered in this course. So we'll use a function which does this for us. `numpy.polyfit()` is one such function. It accepts three main parameters (there are others as well, but they are optional): a list of $x$ coordinates, a list of $y$ coordinates, and a polynomial degree.

There are three main parts to a recursive function:
1. Bottom - when the recursion should finish
2. Operation - some meaningful thing to do
3. Recursive call - calling the same function
4. Clean-up - returning all data to its previous state (this reverses the effect of the operation)

Let's do one of the most famous recursion examples. The most trivial function to recreate is the **factorial**. Derive (or look up) a recurrent definition of the factorial of a non-negative integer and implement it.

**There is a one-to-one correspondence between loops and recursion.** Use this correspondence to create an iterative definition (one which uses loops) and compare both. Test that they return the same results.

**Note:** Think how Python handles very large integers.

````python
def factorial_recursive(n):
    """
    Computes the factorial of a non-negative number n (using recursion)
    """
    if n == 0:
        return 1
    
    # 5! = 5 * 4 * 3 * 2 * 1 = 5 * 4! = 5 * 4 * 3! = 5 * 4 * 3 * 2! 
    return n * factorial_recursive(n - 1)
    

def factorial_iterative(n):
    """
    Computes the factorial of a non-negative number n (using iteration)
    """
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
````

Use the approach above to rewrite the factorial function in a "lazy" way.

````python
def factorial_new(n):
    """ This function calculates the factorial sequence up to the nth term lazily using a generator.
    The factorial sequence is a series of numbers where each term is the factorial of its index. """
    result = 1
    index = 1
    while index <= n:
        result *= index
        yield result
        index += 1

factorial_generator = factorial_new(5)
print(list(factorial_generator))  
````

Use **the same approach** to return the Fibonacci sequence - another famous one in mathematics.

````python
def fibonacci(n):
    """ This function calculates the Fibonacci sequence up to the nth term using recursion.
    The Fibonacci sequence is a series of numbers in which each number (Fibonacci number)
    is the sum of the two preceding ones, starting with 0 and 1. """

    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib_sequence = fibonacci(n - 1)
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
        return fib_sequence
````

#### The Mandelbrot set
Now, let's get to something more beefy. The Mandelbrot set is defined by repeatedly squaring complex numbers. It turns out that it produces a **fractal** - a self-similar structure. Follow a tutorial to create (and possibly visualize) the Mandelbrot set. I think [this one](https://medium.com/swlh/visualizing-the-mandelbrot-set-using-python-50-lines-f6aa5a05cf0f) is a good starting point.

````python
def in_mandelbrot(c, max_iter):
    z = 0
    for i in range(max_iter):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return max_iter

# Define the parameters for the Mandelbrot set
xmin, xmax, ymin, ymax = -2.0, 2.0, -2.0, 2.0
width, height = 800, 800
max_iter = 100

# Create the grid of complex numbers
x = np.linspace(xmin, xmax, width)
y = np.linspace(ymin, ymax, height)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Compute whether each point is in the Mandelbrot set
mandelbrot_set = np.zeros((height, width))
for i in range(height):
    for j in range(width):
        mandelbrot_set[i, j] = in_mandelbrot(Z[i, j], max_iter)

# Visualize the Mandelbrot set
plt.figure(figsize=(10, 10))
plt.imshow(mandelbrot_set, extent=(xmin, xmax, ymin, ymax), cmap='hot', interpolation='bilinear')
plt.colorbar(label='Iterations')
plt.title('Mandelbrot Set')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
````

#### * Optional problem
Try to recreate a Lindermeyer system (also known as **L-system**, [source](https://en.wikipedia.org/wiki/L-system)). These are systems created via a series of replacement rules. 

````python
def apply_rules(axiom, rules, iterations):
    result = axiom
    for _ in range(iterations):
        next_result = ""
        for char in result:
            next_result += rules.get(char, char)
        result = next_result
    return result

def plot_l_system(l_system, angle, initial_length=1):
    x = [0]
    y = [0]
    direction = 90  # Start position
    length = initial_length  # Start length 

    for command in l_system:
        if command == "F":
            # Add point
            x.append(x[-1] + length * cos(radians(direction)))
            y.append(y[-1] + length * sin(radians(direction)))
        elif command == "+":
            # Turn to left
            direction += angle
        elif command == "-":
            # Turn to Right
            direction -= angle

    plt.plot(x, y, color='green')
    plt.axis('equal')
    plt.show()

rules = {
    "F": "F+F-F-F+F",  
}

axiom = "F"

iterations = 3

result = apply_rules(axiom, rules, iterations)

plot_l_system(result, angle=90)
````

### Problem 4. Run-length Encoding
One application of algebra and basic math can be **compression**. This is a way to save data in less space than it originally takes. The most basic form of compression is called [run-length encoding](https://en.wikipedia.org/wiki/Run-length_encoding).

Write a function that encodes a given text. Write another one that decodes.

We can see that RLE is not very useful in the general case. But it can be extremely useful if we have very few symbols. An example of this can be DNA and protein sequences. DNA code, for example, has only 4 characters.

Test your encoding and decoding functions on a DNA sequence (you can look up some on the Internet). Measure how much your data is compressed relative to the original.

````python
def encode(text):
    """
    Returns the run-length encoded version of the text
    (numbers after symbols, length = 1 is skipped)
    """
    encoded_text = ""
    count = 1
    for i in range(1, len(text)):
        if text[i] == text[i-1]:
            count += 1
        else:
            if count > 1:
                encoded_text += text[i-1] + str(count)
            else:
                encoded_text += text[i-1]
            count = 1
    if count > 1:
        encoded_text += text[-1] + str(count)
    else:
        encoded_text += text[-1]
    return encoded_text

def decode(text):
    """
    Decodes the text using run-length encoding
    """
    decoded_text = ""
    i = 0
    while i < len(text):
        char = text[i]
        if char.isdigit():
            decoded_text += text[i-1] * (int(char) - 1)
        else:
            decoded_text += char
        i += 1
    return decoded_text
````

````python
def run_experiment(bit_lengths):
    factorization_times = []
    multiplication_times = []

    for n_bits in bit_lengths:
        factorization_total_time = 0
        multiplication_total_time = 0
        num_trials = 5

        for _ in range(num_trials):
            # Генерираме случайно число
            n = generate_random_number(n_bits)

            # Измерваме времето за факторизация
            start_time = time.time()
            factors = factorize_number(n)
            end_time = time.time()
            factorization_total_time += end_time - start_time

            # Измерваме времето за умножение на факторите
            start_time = time.time()
            result = 1
            for factor, power in factors.items():
                result *= factor ** power
            end_time = time.time()
            multiplication_total_time += end_time - start_time

        # Смятаме средните времена
        factorization_avg_time = factorization_total_time / num_trials
        multiplication_avg_time = multiplication_total_time / num_trials

        # Добавяме ги в списъците
        factorization_times.append(factorization_avg_time)
        multiplication_times.append(multiplication_avg_time)

    return factorization_times, multiplication_times
````

### * Problem 6. Diffie - Hellman Key Exchange Simulation
As we already saw, there are functions which are very easy to compute in the "forward" direction but really difficult (computationally) to invert (that is, determine the input from the output). There is a special case: the function may have a hidden "trap door". If you know where that door is, you can invert the function easily. This statement is at the core of modern cryptography.

However, if you know something else (this is called a **trap door**), the function becomes relatively easy to invert.

````python
def generate_large_prime(bits):
    """
    Generates a large prime number with the specified number of bits.
    """
    prime = sympy.randprime(2 ** (bits - 1), 2 ** bits)
    return prime

def modular_pow(base, exponent, modulus):
    """
    Computes (base ** exponent) % modulus efficiently using modular exponentiation.
    """
    result = 1
    base = base % modulus
    while exponent > 0:
        if exponent % 2 == 1:
            result = (result * base) % modulus
        exponent = exponent >> 1
        base = (base * base) % modulus
    return result

def diffie_hellman_key_exchange(bits):
    """
    Simulates the Diffie-Hellman key exchange process between two participants.
    """
    # Generate large prime numbers p and a generator g
    p = generate_large_prime(bits)
    g = random.randint(2, p - 1)

    # Choose random private keys for both participants
    private_key_A = random.randint(2, p - 2)
    private_key_B = random.randint(2, p - 2)

    # Calculate public keys for both participants
    public_key_A = modular_pow(g, private_key_A, p)
    public_key_B = modular_pow(g, private_key_B, p)

    # Calculate shared secret keys for both participants
    shared_secret_A = modular_pow(public_key_B, private_key_A, p)
    shared_secret_B = modular_pow(public_key_A, private_key_B, p)

    # Verify that both participants have the same shared secret key
    assert shared_secret_A == shared_secret_B

    return shared_secret_A

def run_simulation(bits):
    """
    Runs a simulation of the Diffie-Hellman key exchange and measures the time taken.
    """
    start_time = time.time()
    shared_secret = diffie_hellman_key_exchange(bits)
    end_time = time.time()
    time_taken = end_time - start_time

    return shared_secret, time_taken

bits = 512
shared_secret, time_taken = run_simulation(bits)
print(f"Shared secret: {shared_secret}")
print(f"Time taken: {time_taken} seconds")
````

### ** Huffman Compression Algorithm
Difference Between Lossless and Lossy Compression

**Lossless Compression**:

    Lossless compression reduces the file size without any loss of data. When decompressed, the original data is perfectly restored.
    Examples: Huffman Coding, Lempel-Ziv-Welch (LZW), Run-Length Encoding (RLE).

**Lossy Compression**:

    Lossy compression reduces file size by removing some data, which might result in a loss of quality. It is irreversible.
    Examples: JPEG (for images), MP3 (for audio), MPEG (for video).

**When Can We Get Away with Lossy Compression?**

    Lossy compression is often used in multimedia applications where some loss of quality is acceptable and not easily noticeable by human perception. For instance:
        Images: JPEG
        Audio: MP3
        Video: MPEG

**What is Entropy?**

    Entropy in information theory, introduced by Claude Shannon, is a measure of the average information content per message in a stream of data. It quantifies the uncertainty involved in predicting the value of a random variable.
    In the context of compression, higher entropy implies more randomness and less compressibility, while lower entropy implies more redundancy and higher compressibility.

*Construction of Huffman Trees*

**Steps to Construct Huffman Trees:**

    Frequency Calculation: Calculate the frequency of each character in the data.
    Priority Queue: Insert each character (or symbol) with its frequency into a priority queue (min-heap).
    Tree Construction:
        While there is more than one node in the queue:
            Remove the two nodes with the lowest frequency.
            Create a new node with these two nodes as children, and the frequency as the sum of their frequencies.
            Insert this new node back into the queue.
    Root Node: The remaining node is the root of the Huffman Tree.

**Example:**
**Given a string example with characters and frequencies:**

    e: 2
    x: 1
    a: 1
    m: 1
    p: 1
    l: 1

**Steps:**

    Insert nodes into the priority queue.
    Combine nodes with the lowest frequencies repeatedly until one node remains.

**Decoding Huffman Trees**

*To decode data from a Huffman tree:*

    Start from the root and traverse the tree according to the bits in the encoded data (0 for left, 1 for right).
    When you reach a leaf node, append the character to the decoded output and return to the root to decode the next set of bits.

*Storing Huffman Trees*

    The Huffman tree or codebook can be stored using methods like:
        Explicit Tree Representation: Store the structure of the tree and the characters.
        Implicit Codebook: Store the mappings of characters to their binary codes.

*Implementation*

**Here is an implementation in Python:**


````python


class Node:
    def __init__(self, char=None, freq=None):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None
    
    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(frequencies):
    heap = [Node(char, freq) for char, freq in frequencies.items()]
    heapq.heapify(heap)
    
    while len(heap) > 1:
        node1 = heapq.heappop(heap)
        node2 = heapq.heappop(heap)
        merged = Node(freq=node1.freq + node2.freq)
        merged.left = node1
        merged.right = node2
        heapq.heappush(heap, merged)
    
    return heap[0]

def generate_codes(node, prefix="", codebook=None):
    if codebook is None:
        codebook = {}
    if node:
        if node.char is not None:
            codebook[node.char] = prefix
        generate_codes(node.left, prefix + "0", codebook)
        generate_codes(node.right, prefix + "1", codebook)
    return codebook

def huffman_encoding(data):
    if not data:
        return "", None

    frequency = Counter(data)
    huffman_tree = build_huffman_tree(frequency)
    codebook = generate_codes(huffman_tree)
    encoded_data = ''.join(codebook[char] for char in data)
    return encoded_data, huffman_tree

def huffman_decoding(encoded_data, huffman_tree):
    decoded_data = []
    current_node = huffman_tree
    for bit in encoded_data:
        current_node = current_node.left if bit == '0' else current_node.right
        if current_node.char is not None:
            decoded_data.append(current_node.char)
            current_node = huffman_tree
    return ''.join(decoded_data)

# Test the algorithm
data = "this is an example for huffman encoding"
encoded_data, tree = huffman_encoding(data)
decoded_data = huffman_decoding(encoded_data, tree)

print(f"Original data: {data}")
print(f"Encoded data: {encoded_data}")
print(f"Decoded data: {decoded_data}")

compression_ratio = len(encoded_data) / (len(data) * 8) * 100
print(f"Compression ratio: {compression_ratio:.2f}%")
````

*Huffman Coding vs LZ77:*

    Huffman Coding: Efficient for data with predictable and repetitive patterns. Compression ratio depends on the frequency distribution of characters.
    LZ77: Efficient for larger datasets with more complex redundancy and longer repeated sequences. Typically, it provides better compression for large files.

Huffman Coding is a fundamental algorithm for lossless data compression. It is optimal when the character frequencies are known and do not change.
